# EDGE DETECTION USING CONVOLUTION AND FOURIER ANALYSIS


Importing all the necessary libraries...

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Convert Images to greyscale

```python
new_img = mpimg.imread('C:/Users/ashok/Downloads/cat.png')[:,:,0]
plt.imshow(new_img, cmap = "gray")
plt.savefig("greycat.jpg")
print(new_img)
```

## Generate the sobel filters

```python
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)

xx , yy = np.meshgrid(x, y)
```

```python
conv_img = signal.convolve2d(new_img,  -yy/100, mode = 'same', boundary = 'symm')
print(conv_img)
```

```python
plt.imshow(np.absolute(conv_img * 100000), cmap ='gray')
plt.colorbar()
plt.savefig("cat_edge.png")
```

```python
from scipy import fft
fouri = np.fft.fftshift(fft.fft2(new_img))
print(fouri)
plt.imshow(100*np.log(abs(fouri)), cmap = "gray")
plt.savefig("fouricat.png")
```

## DISCRETE FOURIER TRANSFORMS


```python
import numpy as np
import matplotlib.pyplot as plt
image_filename = "C:/Users/ashok/Downloads/cat.png"
def calculate_2dft(input):
    
    #ft = np.fft.ifftshift(input)
    """ft =""" 
    ft = np.fft.fft2(input)
    return np.fft.fftshift(ft)
# Read and process image
image = plt.imread(image_filename)
plt.figure()
plt.subplot(1, 2, 1)
#plt.subplot()
plt.imshow(image)
image = image[:, :, :3].mean(axis=2)
print(image)# Convert to grayscale
plt.set_cmap("gray")
ft = calculate_2dft(image)
print(np.log(abs(ft)))

plt.subplot(1,2,2)

plt.imshow(10 * np.log(abs(ft)), cmap = "gray")
plt.axis("on")
plt.savefig('cat_fouri.png')
plt.show()

```
